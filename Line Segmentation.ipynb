```python
import cv2
import math
import numpy as np
import os
import glob
import imageio

path1 = os.getcwd()
directory = "out"
path2 = os.path.join(path1, directory)
os.mkdir(path2)
path = path1 + '/out/'

def createKernel(kernelSize, sigma, theta):
    assert kernelSize % 2 # must be odd size
    halfSize = kernelSize // 2
    kernel = np.zeros([kernelSize, kernelSize])
    sigmaX = sigma
    sigmaY = sigma * theta

    for i in range(kernelSize):
        for j in range(kernelSize):
            x = i - halfSize
            y = j - halfSize
            expTerm = np.exp(-x**2 / (2 * sigmaX) - y**2 / (2 * sigmaY))
            xTerm = (x**2 - sigmaX**2) / (2 * math.pi * sigmaX**5 * sigmaY)
            yTerm = (y**2 - sigmaY**2) / (2 * math.pi * sigmaY**5 * sigmaX)
            kernel[i, j] = (xTerm + yTerm) * expTerm

    kernel = kernel / np.sum(kernel)
    return kernel

#binary
Total = 0
for image_path in glob.glob(path1 + "/test\\*.jpg"):
    Total+=1
    direc = Total
    path4 = os.path.join(path2, str(Total))
    os.mkdir(path4)
    image = imageio.imread(image_path)
    #cv2.imshow('image',image)
    #cv2.waitKey(0)
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    kernel = createKernel(1001,11,200)
    imgFiltered = cv2.filter2D(gray, -1, kernel, borderType=cv2.BORDER_REPLICATE).astype(np.uint8)
    ret,thresh = cv2.threshold(imgFiltered,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    thresh = 255 - thresh
    
    #dilation
    kernel = np.ones((5,5), np.uint8)
    img_dilation = cv2.dilate(thresh, kernel, iterations=1)

#find contours
    ctrs, hier = cv2.findContours(img_dilation.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

#sort contours
    sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
    sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[1])
    count = 0
    for i, ctr in enumerate(sorted_ctrs):
    # Get bounding box
        x, y, w, h = cv2.boundingRect(ctr)
        roi = image[y:y+h, x:x+w]
        
    # show ROI
        if w >= 1000 & h>=20 :
            cv2.imwrite(os.path.join(path2,str(Total),"segment_no_" + str(count) + ".png"),roi)
            cv2.rectangle(image,(x,y),( x + w, y + h ),(90,0,255),2)
            count+=1
    #cv2.imwrite('final_bounded_box_image' + str(Total) +'.png',image)
    file = open(path + 'total_line_segments.txt','w')
    file.write("%d" % Total)
    file.close()
```
